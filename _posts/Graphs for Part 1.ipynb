```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
import seaborn as sns
import pandas as pd
import numpy as np
import math
```

```python
gold_x = [0.035274, 1, 35.274]
#gold_y = [i*1313.40 for i in gold_x]
gold_y =[42.38, 1318.25, 42382.72]
gold = pd.DataFrame({"Weight": gold_x, "Price": gold_y})
```

```python
ax = sns.lmplot(x='Weight', y='Price', data=gold)
ax.set(xlabel='Weight of Gold in Ounces (x)', ylabel='Price of Gold (y)')
props = dict(boxstyle='round', alpha=0.5,color=sns.color_palette()[0])
#textstr = "1199.92034887x - 58.37109154"
textstr = 'y=1199.920x - 58.371'
ax.ax.text(0.1, .9, textstr, transform=ax.ax.transAxes, fontsize=14, bbox=props)
plt.show()
#sns.lmplot(x=gold_x, y=gold_y)
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
X = gold.Weight.values.reshape(-1,1)
y = gold.Price.values.reshape(-1,1)
regr.fit(X, y)
print(regr.coef_[0])
print(regr.intercept_)
```

# FLU

```python
flu_x = [i for i in range(20)]
#gold_y = [i*1313.40 for i in gold_x]
eq = lambda x: (7) * (math.e) ** (.2 * x)
flu_y =[eq(i) for i in flu_x]
flu = pd.DataFrame({"Week": flu_x, "Cases": flu_y})
```

```python
ax = sns.lmplot(x='Week', y='Cases', data=flu, order=10, ci=None, scatter_kws={"s": 80})
ax.set(xlabel='Week # (x)', ylabel='Number of Cases (y)')
props = dict(boxstyle='round', alpha=0.5,color=sns.color_palette()[0])
#textstr = "1199.92034887x - 58.37109154"
textstr = '$y = 7e^{.2x}$'
ax.ax.text(0.1, .9, textstr, transform=ax.ax.transAxes, fontsize=14, bbox=props)
#sns.lmplot(x=gold_x, y=gold_y)
```
